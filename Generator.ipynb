#Generator and Yield

```python
def my_gen():
  n = 1
  print('This is printed first')
  yield n 

  n += 1
  print('This is printed second')
  yield n

  n += 1
  print('This is printed at last')
  yield n
```

```python
a = my_gen()
print(next(a))
```

```python
def fibonacci_numbers(nums):
  x, y = 0, 1
  for _ in range(nums):
    x, y = y, x+y
    yield x
```

```python
fn = fibonacci_numbers(10)

for i in fn:
  print(i)
```

```python
def mygen():
  n = 1
  print(n)
  yield n

  n += 1
  print(n)
  yield n

  n += 1
  print('last',n)
  yield n

a2 = mygen()
la2 = list(a2)
print(la2)
```

```python
names = ['Alice', 'Bob', 'Chris', 'David', 'Emily']

def gen_roster(names):
    for name in names:
        yield name
```

```python
rs = gen_roster(names)
print(next(rs))
```

```python
def odd(numbers):
  for number in range(numbers):
    if(number%2 != 0):
      yield number
    
odd_1 = odd(10)
print(list(odd_1))
```

```python
def square(num):
  for nums in range(num):
    if nums%2 == 1:
     yield nums*nums
sq = square(10)

print(list(sq))
```

```python
import math 
import matplotlib.pyplot as plt

def gen_sin():
  t = 0 
  omega = 0.01
  while(True):
    t += 0.1
    omega += 0.001
    yield math.sin(omega*t)
  
t = 0
ygen = gen_sin()
plt.axis([0,10,0,1])
for i in range(100):
  t += 0.1
  y = next(ygen)

  plt.plot(t,y,'o',color='black');
  plt.pause(0.05)

plt.show()
```
